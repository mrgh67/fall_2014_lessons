```python
def perDiff(a, b):
    sameCount = 0
    diffCount = 0
    aL = len(a)
    bL = len(b)
    
    if aL > bL:
        diffCount = aL - bL
        n = bL
    else:
        diffCount = bL - aL
        n = aL
    for i in range(0,n):
        if a[i] == b[i]:
            sameCount = sameCount + 1
        else:
            diffCount = diffCount + 1
    return(diffCount)
```

```python
def wordDiff(a, b):
    diffCount = 0
    s1 = a.split(' ')
    s2 = b.split(' ')
    #print s1
    #print s2
    aS = len(s1)
    bS = len(s2)      
    #print aS, bS
    if aS > bS:
        diffCount = 10 * (aS - bS)
        nn = bS
    else:
        diffCount = 10 * (bS - aS)
        nn = aS
    #print "nn is " , nn
    for i in range(0, nn):
            if(s1[i]!=s2[i]):
                #print s1[i], s2[i]
                diffCount = diffCount + perDiff(s1[i], s2[i])
    return diffCount
```

```python
def accuracy_report(_clf, xtrain,ytrain,xtest, ytest):
    print "Accuracy: %0.2f%%" % (100 * _clf.score(xtest, ytest))

    #Print the accuracy on the test and training dataset
    training_accuracy = _clf.score(xtrain, ytrain)
    test_accuracy = _clf.score(xtest, ytest)

    print "Accuracy on training data: %0.2f" % (training_accuracy)
```
